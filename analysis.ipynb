```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
# setting a default theme for plot
plt.figure(figsize=(10,6))
sns.set(style="whitegrid")
```

```python
df=pd.read_csv("demand_forecasting.csv")
df.head(3)
```

```python
df.dtypes
```

```python
# converted the data to pandas date format
df["Date"]=pd.to_datetime(df["Date"]) 
```

```python
df.dtypes
```

```python
df.info()
```

```python
# check null value 
df.isna().sum()
```

```python
# check for duplicate value in features
df.duplicated().sum()
```

```python
# return the transpose of describe table 
df.describe().T
```

```python
df.describe(include="object").T
```

```python
df.head(2)
```

```python
# create 2 new feature in DataFrame (Year & Month)
# copy year from df["Year"] feature of the table
df["Year"]=df["Date"].dt.year 
df["Month"]=df["Date"].dt.month
df["Day"]=df["Date"].dt.day
df["Weekend"]=df["Date"].dt.day_name()
df.head(2)
```

```python
df.columns
```

```python
# calculate the discounted price
df["Discounted_Price"]=df["Price"]*(1-df["Discount"]/100)
```

```python
df["Sell Through Rate"]=df["Units Sold"] / df["Inventory Level"]
```

```python
df.head(3)
```

```python
df["Sell Through Rate"].describe()
```

```python
df.groupby("Category")["Demand"].agg(["mean","sum","std"]).sort_values(by = "sum", ascending=False).reset_index()
```

```python
df.groupby(["Region","Seasonality"])["Demand"].mean()
```

```python
df.groupby("Promotion")["Demand"].mean()
```

```python
pd.pivot_table(df,values="Demand",index="Month",columns="Category",aggfunc="mean")
```

```python
sns.histplot(df["Demand"],bins=20,kde=True)
plt.title("Demand Distribution")
plt.show()
```

```python
sns.scatterplot(data=df, x = "Inventory Level",y="Units Sold")
plt.title("Inventory Vs Unit Sold")
plt.show()
```

```python
sns.boxplot(data=df,x="Category",y="Demand")
plt.xticks(rotation =45)
plt.title("Demand by Category")
plt.show()
```

```python
sns.boxplot(data=df,x="Weather Condition",y="Demand")
plt.xticks(rotation=45)
plt.title("Demand by Weather Condition")
plt.show()
```

```python
monthly_demand=df.groupby("Month")["Demand"].mean()
```

```python
monthly_demand
```

```python
monthly_demand.plot(kind="bar")
plt.title("Average Demand by Month")
plt.show()
```

```python
daily_demand=df.groupby("Date")["Demand"].sum()
daily_demand.plot()
plt.title("Total Daily Demand Over Time")
plt.xlabel("data")
plt.ylabel("Demand")
plt.show()
```

```python
sns.barplot(data=df,x="Promotion",y="Demand")
plt.title("Promotion Impack on Demand")
plt.show()
```

```python
sns.scatterplot(data=df,x="Discounted_Price",y="Demand")
plt.title("Discounted Price vs Demand")
plt.show()
```

```python
df.groupby("Seasonality")["Demand"].mean().plot(kind="bar",title="Demand By seasonalitye")
plt.show()
```

```python
df.groupby("Epidemic")["Demand"].mean().plot(kind="bar",title="Epidemic impact on Demand")
plt.show()
```
